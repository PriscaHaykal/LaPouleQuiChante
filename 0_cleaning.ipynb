On ne conserve que les colonnes utiles pour la suite soit les colonnes années et pays 

```python
population = population[['Code zone','Zone','Année','Valeur']]
```

```python
population['Valeur']=population['Valeur']*1000
```

```python
population = population.rename(columns={'Code zone':'code_pays','Zone':'pays','Année':'annee','Valeur':'population'})
population.head()
```

On réalise un pivot pour avoir les pays en lignes et les années en colonne 

```python
population = pd.pivot_table(population, values='population',index=['code_pays','pays'],columns='annee')
```

```python
population.head()
```

```python
population[population.isna().any(axis=1)]
```

On supprime les pays qui n'existaient pas avant 2011 ainsi que Serbie-et-Monténégro

```python
population = population.drop(population[(population.index.get_level_values('code_pays') == 186)
                                        |(population.index.get_level_values('code_pays') == 278)
                                        |(population.index.get_level_values('code_pays') == 279)
                                        |(population.index.get_level_values('code_pays') == 280)
                                        |(population.index.get_level_values('code_pays') == 281)
                                        |(population.index.get_level_values('code_pays') == 282)].index)
```

```python
population[population.isna().any(axis=1)]
```

```python
pop_soudan_total_2012 = population[population.index.get_level_values('code_pays')==276][2012].values[0] + population[population.index.get_level_values('code_pays')==277][2012].values[0]
part_soudan = population[population.index.get_level_values('code_pays')==276][2012].values[0]/pop_soudan_total_2012
part_soudan_du_sud = population[population.index.get_level_values('code_pays')==277][2012].values[0]/pop_soudan_total_2012

print("Le soudan représente {}% de la population totale du soudan et le soudan du Sud représente {}%".format(part_soudan*100,part_soudan_du_sud*100))
```

```python
population.loc[276,2007] = part_soudan*population[population.index.get_level_values('code_pays')==206][2007].values[0]
#df.loc[df[<some_column_name>] == <condition>, [<another_column_name>]] = <value_to_add>
population.loc[276, :]
```

```python
population.loc[277,2007] = part_soudan_du_sud*population[population.index.get_level_values('code_pays')==206][2007].values[0]
population.loc[277, :]
```

On supprime la ligne Soudan (ex)

```python
population = population.drop(population[(population.index.get_level_values('code_pays') == 206)].index)
```

On crée une colonne pour l'évolution de la population entre 2007 et 2017

```python
population['pop_evolution_ratio'] = (population[2017] - population[2007])/population[2007]
population.head()
```

On garde les colonnes utiles pour la suite

```python
population = population[[2017,'pop_evolution_ratio']]
population = population.rename(columns = {2017:'pop_2017'})
population.head()
```

```python
population = population.reset_index()
```

# Etude du fichier Dispo_Viande_Volailles_2017

```python
dispo_volailles = pd.read_csv('data/Dispo_Viande_Volailles_2017.csv')
dispo_volailles.head()
```

```python
dispo_volailles.info()
```

```python
dispo_volailles = pd.pivot_table(dispo_volailles, values='Valeur',index=['Code zone (ISO3)',"Zone"],columns='Élément')
dispo_volailles.head()
```

```python
dispo_volailles = dispo_volailles.reset_index()
dispo_volailles.head()
```

```python
dispo_volailles = dispo_volailles.rename(columns={"Code zone (ISO3)":"code_iso_pays" , 
                                                  "Zone" : "pays",
                                                  "Disponibilité alimentaire (Kcal/personne/jour)" : "dispo_volailles_kcal_pers_jour",
                                                  "Importations - Quantité" :"quantite_importee",
                                                  "Nourriture" : "nourriture",
                                                  "Production" : "quantite_produite"})
dispo_volailles.head()
```

```python
dispo_volailles['quantite_importee'] = dispo_volailles['quantite_importee']*10e+5
dispo_volailles['quantite_produite'] = dispo_volailles['quantite_produite']*10e+5
dispo_volailles['nourriture'] = dispo_volailles['nourriture']*10e+5
dispo_volailles.head()
```

```python
dispo_volailles = dispo_volailles.rename(columns = {'quantite_importee':'quantite_importee_kg',
                                                    'quantite_produite':'quantite_produite_kg',
                                                    'nourriture':'nourriture_kg'})
dispo_volailles.head()
```

```python
dispo_volailles.info()
```

```python
dispo_volailles[dispo_volailles.isna().any(axis=1)]
```

On observe qu'il y a des valeurs manquantes pour 3 pays : Djibouti, Maldives et République populaire démocratique de Corée.
Je décide de ne pas les inclure dans mon analyse.

```python
dispo_volailles.drop(dispo_volailles.loc[dispo_volailles['pays']=='Djibouti'].index, inplace=True)         
dispo_volailles.drop(dispo_volailles.loc[dispo_volailles['pays']=='Maldives'].index, inplace=True)
dispo_volailles.drop(dispo_volailles.loc[dispo_volailles['pays']=='République populaire démocratique de Corée'].index, inplace=True)
```
